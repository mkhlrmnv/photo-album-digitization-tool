```python
from ultralytics import YOLO
from roboflow import Roboflow
import cv2
import os
import glob
from dotenv import load_dotenv
load_dotenv() 
```

```python
'''
Block for preprocessing images, before putting them into roboflow.
basically we binary black and white each image, to make training easier
'''

image_extension = "*.jpeg"
image_folder = "/Users/mkhlrmnv/Desktop/kuva-prosessointi/approach_with_custom_ml/with_pre_processing/data_set/orig"

output_folder = "/Users/mkhlrmnv/Desktop/kuva-prosessointi/approach_with_custom_ml/with_pre_processing/data_set/processed"

imgs = []
imgs.extend(glob.glob(os.path.join(image_folder, image_extension)))

for image in imgs:
    img = cv2.imread(image)
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray_img, 240, 255, cv2.THRESH_BINARY_INV)

    img_name = image.split("/")[-1]

    print(img_name)

    new_name = os.path.join(output_folder, img_name)

    cv2.imwrite(new_name, thresh)


```
